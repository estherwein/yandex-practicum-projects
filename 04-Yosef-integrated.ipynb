### rating

the rating is an ordinal variable. we will use it to fill the missing values based on the median value.
we first will give rating column a numerical value, then replace the missing values with the median.

````python
df_games["rating"].value_counts()
````

from search on the internet "k-a" is "kids to adoults" and become "e", "ec" is "early childhood" and removed too and became "e".
rp is return pending, so we replace it with null.       
https://www.esrb.org/history/       
https://rating-system.fandom.com/wiki/ESRB_Early_Childhood       

````python
def map_values(x):
    if x=="k-a" or x=='ec':
        return "e"
    elif x=="rp":
        return np.nan
    else: 
        return x
df_games["rating"] = df_games["rating"].apply(map_values)
````

````python
df_games["rating"].value_counts()
````

we have now 5 ordinal values, we will replace them with numbers, and fill the null values with the median:

````python
# create new numerical rating column (0 for every body, 5 adoults only)
cat_list = ["e","e10+","t","m","ao"]
cat_dict = dict(zip(cat_list, np.arange(len(cat_list))))
df_games['rating_numerical'] = df_games['rating'].map(cat_dict)
# fill the null values with the median:
df_games['rating_numerical'] = df_games['rating_numerical'].fillna(df_games['rating_numerical'].median())
````

````python
df_games.head()
````

we see that the median is 1 ("e10+") and we replaced the nulls to it.

````python
df_games['rating_numerical'].value_counts()
````

### Null : 
$\Large\mu$$_{xbox-user-rating}$ - $\Large\mu$$_{pc-user-rating}$ =0 

    
### Alternative: 
**$\Large\mu$$_{xbox-user-rating}$** - **$\Large\mu$$_{pc-user-rating}$** != 0    

$\Large\alpha$ = $\Large5\%$

### the null hypothesis suggests that there isn't a difference between the platforms average user rating, the alternative suggests there is a difference.
### we choose alpha 5% because it's a standard.

````python
from scipy import stats as st
import numpy as np

alpha = 0.05  # critical statistical significance level
# if the p-value is less than alpha, we reject the hypothesis
arr1= relevant.query("platform=='xone'")["user_score"]
arr2= relevant.query("platform=='pc'")["user_score"]

results = st.ttest_ind(arr1, arr2,equal_var=False)

print('p-value: ', results.pvalue)

if results.pvalue < alpha:
    print("We reject the null hypothesis")
else:
    print("We can't reject the null hypothesis") 
````

    that there is a difference between the platforms.

### test: ‚ÄîAverage user ratings for the Action and Sports genres are different.

### Null : 
$\Large\mu$$_{action-user-rating}$ - $\Large\mu$$_{sports-user-rating}$ =0 

    
### Alternative: 
**$\Large\mu$$_{action-user-rating}$** - **$\Large\mu$$_{sports-user-rating}$** != 0    

$\Large\alpha$ = $\Large5\%$

### the null hypothesis suggests that there isn't a difference between the genres average rating, the alternative suggests there is a difference.
### we choose alpha 5% because it's a standard.

````python
from scipy import stats as st
import numpy as np

alpha = 0.05  # critical statistical significance level
# if the p-value is less than alpha, we reject the hypothesis
arr1= relevant.query("genre=='action'")["user_score"]
arr2= relevant.query("genre=='sports'")["user_score"]

results = st.ttest_ind(arr1, arr2,equal_var=False)

print('p-value: ', results.pvalue)

if results.pvalue < alpha:
    print("We reject the null hypothesis")
else:
    print("We can't reject the null hypothesis") 
````

      we see that p value is small more than alpha, so we reject the null hypothesis, and we believe that there is a differnce between the user ratings of the games in the two genres.
