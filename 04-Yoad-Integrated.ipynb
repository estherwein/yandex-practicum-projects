### Sales by platform
<a class="anchor" id="chapter18"></a>
[Go back to the Table of Contents](#table_of_contents)

- Here we will examine the differences in sales from different platforms.
- We will identify the platforms with the highest total sales.
- We will take a closer look at how these platforms have performed over time.
- We'll identify the platforms that were previously successful but now have no sales.
- We'll determine the general time it takes for a platform to fade and for a new one to arrive.

To determine which platform is most profitable, we need to: 
1. Group by platform and sort by their 'total_sales'.
2. Calculate the z-score of total sales of each platform, which indicates how far the data is from the average total sales.
3. After that, we will color code z-scores above and below the mean in order to make plotting the graph easier.
4. We will plot a horizontal line graph representing each platform, based on its z-score and color coded accordingly.

Grouping the data by platform and sorting by 'total_sales'.

```python
sales_per_platform = df[['platform', 'total_sales']].groupby(['platform']).sum().sort_values(by='total_sales').reset_index()
```

Calculating the z-score of total sales for each platform.

```python
sales_per_platform['sales_z'] = (sales_per_platform['total_sales']-
                                 sales_per_platform['total_sales'].mean()) / sales_per_platform['total_sales'].std()
```

With the z-score, we know that if the data is normally distributed (statistically speaking), then the three-sigma rule holds true and 99.7% of the data will fall within this range.<br>
Accordingly, the nearer to a value +3 it gets, the higher it is distributed in relation to the rest of the data, and vice versa.<br>
This is how we'll evaluate the success of each platform.

Color coding the platforms, z-score above the mean is green, and below will be red.

```python
sales_per_platform['color'] = ['red' if x<0 else 'green' for x in sales_per_platform['sales_z']]
```

Plotting the graph.

```python
plt.figure(figsize=(12,12))
plt.hlines(y=sales_per_platform['platform'], 
           xmin=0, xmax=sales_per_platform['sales_z'], 
           color= sales_per_platform['color'], alpha=0.4, linewidth=10);
plt.title( 'Platform Sales Z-Score, Comparing Average Platform Sales' , size = 16 )
plt.ylabel('Platform', size=14);
plt.xlabel('Z-Score', size=14);
plt.tick_params(axis='both', which='major', labelsize=12)
```

- Here we will examine the differences in sales from different platforms after 2013.
- We will identify the platforms with the highest total sales.

- We'll select a few promising platforms for our prognosis

To determine which platform is most profitable after 2013, we need to: 
1. Filter the data to have only 2013 and beyond.
2. Group by platform and sort by their 'total_sales'.
3. Calculate the z-score of total sales of each platform, which indicates how far the data is from the average total sales.
4. After that, we will color code z-scores above and below the mean in order to make plotting the graph easier.
5. We will plot a horizontal line graph representing each platform, based on its z-score of total sales and color coded accordingly.

Filtering the data by year 2013 and beyond.

```python
df_2013 = df.query('year_of_release >= 2013')
```

Grouping the data by platform and sorting by 'total_sales'.

```python
sales_per_platform_2013 = df_2013[['platform', 'total_sales']].groupby(['platform']).sum().sort_values(by='total_sales').reset_index()
```

Calculating the z-score of total sales for each platform.

```python
sales_per_platform_2013['sales_z'] = (sales_per_platform_2013['total_sales']-
                                 sales_per_platform_2013['total_sales'].mean()) / sales_per_platform_2013['total_sales'].std()
```

With the z-score, we know that if the data is normally distributed (statistically speaking), then the three-sigma rule holds true and 99.7% of the data will fall within this range.<br>
Accordingly, the nearer to a value +3 it gets, the higher it is distributed in relation to the rest of the data, and vice versa.<br>
This is how we'll evaluate the success of each platform.

Color coding the platforms, z-score above the mean is green, and below will be red.

```python
sales_per_platform_2013['color'] = ['red' if x<0 else 'green' for x in sales_per_platform_2013['sales_z']]
```

Plotting the graph.

```python
plt.figure(figsize=(12,6))
plt.title( 'Platform Sales Z-Score, Comparing Average Platform Sales' , size = 16 )
plt.hlines(y=sales_per_platform_2013['platform'], 
           xmin=0, xmax=sales_per_platform_2013['sales_z'], 
           color= sales_per_platform_2013['color'], alpha=0.4, linewidth=10);
plt.ylabel('Platform', size=14);
plt.xlabel('Z-Score', size=14);
plt.tick_params(axis='both', which='major', labelsize=12)
```

Create a function that, for a given console, does the following:<br> 
- Displays the correlation between user score and sales and critic score and sales.
- Making scatter plots comparing critic and user ratings to total sales.


This function is used in a loop for every other popular platform found in our data.

```python
def score_effect(data,platform):
    # Creating a class for text styling
    class color:
        BOLD = '\033[1m'       # Bold text
        UNDERLINE = '\033[4m'  # Undelined text
        END = '\033[0m'        # End of text styling
    print(color.BOLD + color.UNDERLINE + 'Score / Sales Affect For', platform + color.END)
    print()
    
    # Filtering the data by platfrom
    data = data.query('platform == @platform')
    
    # Filtering data for critic reviews
    data_critic = data.drop('user_score',axis=1)
    print('There are', len(data_critic), 'reviews by critics for that platform')
    print('The correlation between critic score and sales is {:.2f}'
          .format(data_critic['total_sales'].corr(data_critic['critic_score'])))
    print()

    # Filtering data for user reviews
    data_user = data.drop('critic_score',axis=1)
    print('There are', len(data_user), 'reviews by users for that platform')
    print('The correlation between user score and sales is {:.2f}'
          .format(data_user['total_sales'].corr(data_user['user_score'])))
    print()
    
    # Plotting scatter for users and critics review for that platform
    fig, axes = plt.subplots(1, 2, figsize=(12,6), sharex=True)                               # Creating the figure and the axes for plotting.
    fig.suptitle('How User / Critic Score Affect Sales on '+platform, size=25)                # Creating the main title of the subplot.
    sns.set_style("whitegrid")                                                                # Setting the style of the plot to white grid.
    sns.scatterplot(ax=axes[0], x=data_user['user_score'], y=data_user['total_sales']);       # Plotting the first scatter for user score for that platform.
    axes[0].set_title('User Score' , size = 17 );                                             # Setting title of the first graph (left).
    axes[0].set_xlabel('Score', size=15);                                                     # Setting title and size of x axis.
    axes[0].set_ylabel('Total sales in mil.', size=15);                                       # Setting title and size of y axis.

    sns.scatterplot(ax=axes[1], x=data_critic['critic_score'], y=data_critic['total_sales']); # Plotting the second scatter for critic score for that platform.
    axes[1].set_title('Critic Score' , size = 17 );                                           # Setting title of the first graph (left).
    axes[1].set_xlabel('Score', size=15);                                                     # Setting title and size of x axis.
    axes[1].set_ylabel('Total sales in mil.', size=15);                                       # Setting title and size of y axis.
    sns.despine()                                                                             # Styling figure without top and right borders.
    plt.show()                                                                                # Showing plot and preventing plot to apear after the text for all iterations has been printed!
    print()
for x in lead_platforms_2013_other:                                                           # Create a loop to print all of the above for each of the popular platforms.
    score_effect(data=lead_platforms_2013_data_score, platform=x)
```
